```python
import matplotlib.pyplot as plt

print("setup terminado")
```

```python
plt.figure(figsize = (12, 6)) # (ancho, largo) de la grafica

marcas_clase = [23, 26, 29, 32, 35] # datos del eje x
frecuencia = [2, 3, 7, 11, 5] # datos del eje y

# Histograma con:
# - barras ajestadas al 100% (width = 1)
# - contorno negro (edgecolor = k)
plt.bar(marcas_clase, frecuencia, 
        width = 1, edgecolor = "k", 
        color = ["#F78181", "#F5DA81", "#58ACFA", "#BCF5A9", "#F5D0A9", "#BCA9F5"]) 

plt.xlabel("Marcas de Clase", fontsize = 25) # etiqueta del eje x
plt.ylabel("Frecuancia Absoluta", fontsize = 25) # etiqueta del eje y
plt.title("Histograma", fontsize = 40) # titulo de la grafica
plt.grid() # se activa la cuadricula
plt.show() # se muestra el grafico en pantalla
```
