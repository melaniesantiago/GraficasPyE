```python
import matplotlib.pyplot as plt

print("setup terminado")
```

```python
plt.figure(figsize = (12, 6)) # (ancho, largo) de la grafica

marcas_clase = [22.25, 24.295, 25.895, 27.495, 29.095, 30.695, 32.295, 33.9] # datos del eje x
frecuen = [3, 5, 6, 8, 9, 10, 6, 4] # datos del eje y

# Histograma con:
# - barras ajestadas al 100% (width = 1)
# - contorno negro (edgecolor = k)
plt.bar(marcas_clase, frecuen, 
        width = 1, edgecolor = "k", 
        color = ["#F78181", "#F5DA81", "#58ACFA", "#BCF5A9", "#F5D0A9", "#BCA9F5"]) 

marcas_texto = ["22.25", "24.295", "25.895", "27.495", "29.095", "30.695", "32.295", "33.9"]
plt.xticks(marcas_clase , marcas_texto, fontsize= 15, rotation= 45)

plt.xlabel("Temperatura (°C)", fontsize = 25) # etiqueta del eje x
plt.ylabel("# Días", fontsize = 25) # etiqueta del eje y
plt.title("Histograma", fontsize = 40) # titulo de la grafica
plt.grid(True) # se activa la cuadricula
plt.show() # se muestra el grafico en pantalla
```

```python

plt.figure(figsize = (12, 6)) # Ancho, largo del grafico
marcas_clasess = [22.7, 24.9, 27.1, 29.3, 31.5, 33.7]#Datos del eje X
# frecuencia =[0.5, 0.15, 0.20, 0.20, 0.10, 0.30,] #Datos del eje Y
frecuenciass =[0.031, 0.187, 0.406, 0.656, 0.844, 1.000] #Datos del eje Y

# Ajuste en las listas para generar el poligino de frecuencia
datos_x = [0] + marcas_clasess 
datos_y = [0] + frecuenciass 

#Leyendas para las marcas de clase
marcas_textosss = ["22.7", "24.9", "27.1", "29.3", "31.5", "33.7"]


# Poligono de frecuencia con:
# - Estilo de line puntedo ( linewidth = 5, color = "r", linestyle = ":"). https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html
# - Marcador con forma triangular (marker = "v"). https://matplotlib.org/stable/api/markers_api.html
plt.plot(datos_x, datos_y,
        linewidth = 5, color = "r", linestyle = "-",
        marker = "s", markersize = 15, markeredgecolor = "b", markerfacecolor = "y")

plt.xticks(marcas_clasess , marcas_textosss, fontsize= 15, rotation= 45)# Cambio de las marcas de clese del eje x de número a texto.
plt.title("Ojiva", fontsize = 40) # Titulo
plt.xlabel("Temperatura °C", fontsize = 25) # Etiqueta eje X
plt.ylabel("Frecuencia relativa acumulada", fontsize = 25) # Etiqueta eje Y
plt.grid(True)#Cuatricula
plt.show()#Se muestra gráfica en pantalla
```
